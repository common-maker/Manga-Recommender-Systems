```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
from nltk.stem import *
from scipy.stats import pearsonr
from sklearn.preprocessing import MinMaxScaler
```

```python
manga = pd.read_json('data/raw/manga.json')
```

```python
old_size = manga.shape[0]
manga = manga.drop_duplicates(subset=['mal_id']).reset_index(drop=True)
print('Duplicates:', old_size - manga.shape[0])
```

```python
data=manga.sample(n=10000,random_state=42)
```

```python
print(data.shape)
```

```python
print(data.dtypes)
```

```python
data=data.drop(["demographics","url","images","approved","titles","title_english","title_japanese","title_synonyms","scored","authors","explicit_genres","themes","background","synopsis"],axis=1)
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.head(10)
```

```python
data['published_from'] = pd.to_datetime(data['published'].str['from']).dt.date
data['published_to'] = pd.to_datetime(data['published'].str['to']).dt.date
data.drop(columns=['published'], inplace=True)
```

```python
def extract_genre_names(x):
    if isinstance(x, list):
        return ",".join([d['name'] for d in x if 'name' in d])
    else:
        return ""  # empty string for missing or invalid

data['genres'] = data['genres'].apply(extract_genre_names)
data['serializations'] = data['serializations'].apply(extract_genre_names)
```

```python
missing_per_col = data.isna().sum()
print(missing_per_col)
```

Xu ly serialization

```python
label_encoder = LabelEncoder()
data['serializations'] = label_encoder.fit_transform(data['serializations'])
```

```python
avg_score = data['score'].mean()
data['score'] = data['score'].fillna(avg_score)
```

```python
avg_score_by=data['scored_by'].mean()
data['scored_by']=data['scored_by'].fillna(avg_score_by)
```

```python
avr_chapter=data['chapters'].mean()
data['chapters']=data['chapters'].fillna(avr_chapter)
```

```python
avr_volumes=data['volumes'].mean()
data['volumes']=data['volumes'].fillna(avr_volumes)
```

```python
avr_ranks=data['rank'].mean()
data['rank']=data['rank'].fillna(avr_ranks)
```

```python
data=data.astype({'chapters':'int64', 'volumes':'int64','scored_by':'int64','score':'int64','rank':'int64'})
data.dtypes
```

```python
missing_per_col = data.isna().sum()
print(missing_per_col)
```
